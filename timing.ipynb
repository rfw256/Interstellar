```python
import numpy as np
import pandas as pd
import os

wlsubj = 139
design_dir = '../../task/Interstellar/main/results/sub-wlsubj%03d' % wlsubj
filename = 'sub-wlsubj%03d_run-07_perception_08-03-22_14-49_trialdata.tsv' % wlsubj

os.listdir(design_dir)
```

```python
cue = .5
iti = 4
delay = 11.5
sacc = 1.5

n_trials = 16
prescan = 10
postscan = 6

(((cue+iti+delay+sacc) * 16) + prescan + postscan) # + 4.5+cue+delay+sacc
```

```python
file_path = os.path.join(design_dir, filename)
os.path.exists(file_path)
```

```python
timings = pd.read_csv(file_path, sep = '\t', index_col = 0)
timings
```

```python
np.roll(timings.iti_onset, -1)
```

```python
(np.roll(timings.iti_onset, -1) - timings.sacc_onset)[:-1].mean() - 1.5
```

```python
mean_cue_err = (timings.delay_onset - timings.cue_onset).mean() - 0.5
mean_delay_err = (timings.sacc_onset - timings.delay_onset).mean() - 11.5
mean_iti_err = (timings.cue_onset - timings.iti_onset).mean() - (16*4 + 4.5) / 17
mean_sacc_err = (np.roll(timings.iti_onset, -1) - timings.sacc_onset)[:-1].mean() - 1.5

run_time = timings.sacc_onset.iloc[-1]  + 6 + 1.5

print('ERRORS:\n\tCue: %0.03f\n\tDelay: %0.03f\n\tITI: %0.03f\n\tSACC: %0.03f\n\nTOTAL RUN TIME: %0.03f' % (mean_cue_err, mean_delay_err, mean_iti_err, mean_sacc_err, run_time))

print('CUM ERRORS:\n\tCue: %0.03f\n\tDelay: %0.03f\n\tITI: %0.03f\n\tSACC: %0.03f\n\nTOTAL RUN TIME: %0.03f' % (mean_cue_err*16, mean_delay_err*16, mean_iti_err*16, mean_sacc_err*16, run_time))

```
